**Exploratory Data Analysis**


---

EDA is the process of exploring, visualizing, and summarizing data before applying any machine learning or statistical modeling.

It’s like getting to know your data:

✅ understanding patterns

✅ spotting anomalies

✅ checking assumptions

✅ identifying relationships between variables

**Why is EDA Important?**

---



1.   **Understand data quality :** Helps find missing values, duplicates, or incorrect data entries.
Detect outliers and anomalies	Outliers can skew models; knowing them helps you

1.  **Detect outliers and anomalies	:** Outliers can skew models; knowing them helps you decide whether to keep or remove them.

1.   **Uncover patterns & relationships :**	Reveals correlations or trends (e.g., does survival depend on age or ticket class?).

2.   **Choose right models & features :**	Guides feature selection, transformation, and the type of model to apply.

1.   **Prevent wrong assumptions :**	Ensures you don’t blindly apply algorithms without knowing what’s inside the data.
2.   **Improve model performance	:** A well-explored dataset usually leads to a better, more accurate model.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot 1: Overall Age distribution (with missing)
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Overall Age Distribution (with missing)')
plt.show()

# Plot 2: Age distribution by Pclass
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Pclass')
plt.show()

# Plot 3: Age distribution by Sex
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title('Age Distribution by Sex')
plt.show()

# Plot 4: Age distribution by Sex and Pclass
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age Distribution by Pclass and Sex')
plt.show()

```



---
**By graphs, i analyzed that Pclass 1 are more prefered by
adults and Pclass 3 are mostly childrens and younger peoples.**

**And From the Histplot i noticed that median inputation is better than mean inputation**

```python
# Fill missing Age values using median of each Sex + Pclass group

for sex in df['Sex'].unique():
    for pclass in df['Pclass'].unique():
        median_age = df[(df['Sex'] == sex) & (df['Pclass'] == pclass)]['Age'].median()
        df.loc[(df['Age'].isnull()) & (df['Sex'] == sex) & (df['Pclass'] == pclass), 'Age'] = median_age
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot Before Handling Outliers')
plt.show()
```

```python
import numpy as np
df['Fare_log'] = np.log1p(df['Fare'])  # log(1 + Fare) to handle zeroes
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare_log'])
plt.title('Fare Boxplot After Log Transformation')
plt.xlabel('Log Transformed Fare')
plt.show()
```



---
**High fare outliers are handled. Tree-based models are less
affected by them, but log transforms help regression models like Linear and Logistic Regression or SVM perform better.**
